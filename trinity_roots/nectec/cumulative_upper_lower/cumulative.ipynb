```
df_text['accumulative_income'] = df_text['total_together'].cumsum()
df_text['total_all'] = df_text['total_together'].sum()

df_text['accumulative_income_percentage'] = df_text['accumulative_income']/df_text['total_all'] * 100
df_text
```

```
# template
import math 
def clean_data(df,comulative_percent,percent_choose,percent_choose_upper,user_choose_percent):
  percent = len(df[df['accumulative_income_percentage'] < comulative_percent])/(len(df)) * 100
  print('percent',percent)
  if percent <= percent_choose:
    df = df[:int(len(df)/100 * percent_choose)]
  else:
    df = df[:int(len(df)/100 * percent_choose_upper)]
    df = df[:math.ceil(len(df) * user_choose_percent / 100)]


  return df
clean_data(df_text,80,20,5,5)
# clean_data(df_text,80,20,30,5)
```
