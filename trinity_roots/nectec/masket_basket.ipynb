```
with pd.ExcelFile('/content/drive/My Drive/work/nectec/dataSet4 (2)-newFromITaaS.xlsx') as reader:
  # print(reader.sheet_names)
   df_two = pd.read_excel(reader, sheet_name='ItemUpdateOct2018')
df_two
```

```
df_all = pd.merge(df_one, df_two, on='Item No', how='inner')
df_all = df_all[['Receipt No','Item No','Price','Quantity','Date','Time','Description','Category']]
df_all
```

```
df_basket = (df_all
          .groupby(['Receipt No', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('Receipt No'))
df_basket
```

```
import datetime
time_start = datetime.datetime.now()
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

basket_sets = df_basket.applymap(encode_units)
time_end = datetime.datetime.now()
print(time_end - time_start)
basket_sets
```

```
frequent_itemsets = apriori(basket_sets, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```
