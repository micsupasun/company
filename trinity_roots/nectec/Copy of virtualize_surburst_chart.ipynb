```
from calendar import monthrange
import plotly.graph_objects as go
def virtualize_surburst_chart(df,start_date,end_date,category_date):
  # start_date = '2018-01-01'
  # end_date = '2019-12-18'
  clean_data = df.groupby(['Receipt No','year','month','date','Date']).count().reset_index()[['Receipt No','year','month','date','Date','Item No']]
  clean_data = clean_data.rename(columns={"Item No": "count_receipt"})
  clean_data

  list_date = []
  for i in range(len(clean_data)):
    list_date.append(f"{clean_data['year'][i]}-{clean_data['month'][i]}-{clean_data['date'][i]}")
  # len(list_date)
  clean_data['Date'] = pd.to_datetime(list_date)
  df_clean_choose = clean_data.sort_values(['Date'], ascending=[True])
  df_clean_choose['week'] = df_clean_choose['Date'].dt.dayofweek
  df_between_date = df_clean_choose[(df_clean_choose['Date'] > start_date) & (df_clean_choose['Date'] <= end_date)]
  df_between_date['quarter'] = df_between_date['Date'].dt.quarter

  if 'year' == category_date:
    df_sum = df_between_date.groupby(['Receipt No',f'{category_date}','count_receipt']).sum().reset_index()[['Receipt No',f'{category_date}','count_receipt']]
    name_sku = []
    category_sku = []
    for i in range(len(df_sum)):
      if df_sum['count_receipt'][i] > 1:
        name_sku.append(f'many_sku_{df_sum["year"][i]}')
        category_sku.append('many_sku')
      else:
        name_sku.append(f'one_sku_{df_sum["year"][i]}')
        category_sku.append('one_sku')
    df_sum['category_sku'] = category_sku
    df_sum['name_sku'] = name_sku
    # df_sum = df_sum[:2000]

    unique_year_list = sorted(df_sum['year'].unique().tolist())
    label = ["total_sku"]
    parent = [""]
    value = [0]
    list_label_two = [f"{unique_year_list[i]}" for i in range(len(unique_year_list))]
    list_parent_two = ["total_sku" for i in range(len(unique_year_list))]
    list_value_two = [0 for i in range(len(unique_year_list))]
    label = label + list_label_two
    parent = parent + list_parent_two
    value = value + list_value_two
    for i in range(len(unique_year_list)):
      unique_sku_in_year = df_sum[df_sum['year'] == unique_year_list[i]]['name_sku'].unique().tolist()
      for j in range(len(unique_sku_in_year)):
        label.append(unique_sku_in_year[j])
        parent.append(f"{unique_year_list[i]}")
        value.append(0)
      year_many_sku = df_sum[(df_sum['year'] == unique_year_list[i]) & (df_sum['category_sku'] == 'many_sku')]
      year_one_sku = df_sum[(df_sum['year'] == unique_year_list[i]) & (df_sum['category_sku'] == 'one_sku')]
      label = label + year_many_sku['Receipt No'].tolist()
      parent = parent + year_many_sku['name_sku'].tolist()
      value = value + year_many_sku['count_receipt'].tolist()
      label = label + year_one_sku['Receipt No'].tolist()
      parent = parent + year_one_sku['name_sku'].tolist()
      value = value + year_one_sku['count_receipt'].tolist()
  # elif 'month' == category_date:
    # print(df_sum)
    # fig =go.Figure(go.Sunburst(
    #     labels =label,
    #     parents=parent,
    #     values =value,
    # ))

    # fig.update_layout(margin = dict(t=0, l=0, r=0, b=0))

    
  return df_between_date
print(virtualize_surburst_chart(df_clean_all[:100],'2018-01-01','2019-12-18','year'))
```
